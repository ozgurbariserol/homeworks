```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.neighbors import NearestNeighbors
import statistics
```

```python
df = pd.read_csv("/restoran-oneri.txt",encoding= 'unicode_escape',
                 names =["Restoran_Kodu","ortam_şıklığı","ortam_temizliği","yemek_kalitesi","hizmet_kalitesi","fiyat_uygunluğu","ulaşım_kolaylığı","araç_park_olanağı"],skiprows =1)
df = df.replace("?", np.nan)
```

```python
for i in df.columns:  
  if i != 'Restoran_Kodu':    
    df[i] = pd.to_numeric(df[i], errors='coerce')
    df[i] = df[i].fillna(df[i].mean())
    df[i] = df[i].astype("int64")
```

```python
print('1 ile 10 arası puanlamanızı yapınız.')
puan1 = int(input('ortam_şıklığı '))
puan2 = int(input('ortam_temizliği '))
puan3 = int(input('yemek_kalitesi '))
puan4 = int(input('hizmet_kalitesi '))
puan5 = int(input('fiyat_uygunluğu '))
puan6 = int(input('ulaşım_kolaylığı '))
puan7 = int(input('araç_park_olanağı ')) 

k = int(input('Kaç tane restoran önerisi istiyorsunuz? (1-15 Arası deger giriniz.)'))

"""
puan1 = 10
puan2 = 3
puan3 = 2
puan4 = 9
puan5 = 4
puan6 = 9
puan7 = 4

k = 5
"""
```

```python
X = df.iloc[:,1:].values
y = df.iloc[:,0].values
neigh = NearestNeighbors(n_neighbors = k,metric = 'cosine')
neigh.fit(X)

nbrs = NearestNeighbors(n_neighbors=k, metric='correlation')
nbrs.fit(X)
```

```python
Cos_Ind =neigh.kneighbors([[puan1,puan2,puan3,puan4,puan5,puan6,puan7]])[1]
Cos_Deger =neigh.kneighbors([[puan1,puan2,puan3,puan4,puan5,puan6,puan7]])[0]
Pear_Ind =nbrs.kneighbors([[puan1,puan2,puan3,puan4,puan5,puan6,puan7]])[1]
Pear_Deger =nbrs.kneighbors([[puan1,puan2,puan3,puan4,puan5,puan6,puan7]])[0]

while True:
  if Cos_Ind[0,0] != Cos_Ind[0,k-1]:
    break
  if Cos_Ind[0,0] == Cos_Ind[0,k-1]:
    k = k+1
    neigh = NearestNeighbors(n_neighbors = k,metric = 'cosine')
    neigh.fit(X)

    nbrs = NearestNeighbors(n_neighbors=k, metric='correlation')
    nbrs.fit(X)

    nb = neigh.kneighbors([[puan1,puan2,puan3,puan4,puan5,puan6,puan7]])
    Cos_Ind =nb[1]
    Cos_Deger =nb[0]
    Pear_Ind =nb[1]
    Pear_Deger =nb[0]
```

```python
print('Cosinuse Gore Sonuclar')
for i in range(k):
  print('Onerilen {2}.Restorantın Kodu {0}\t,Cosinus Benzerlik Degeri {1}'.format(y[Cos_Ind[0, i]],1-Cos_Deger[0,i],i+1))
  
   
print('\nPearson korelasyon katsayısına gore sonuclar')
for i in range(k):
  print('Onerilen {2}.Restorantın Kodu {0}\t,Pearson korelasyon katsayısı değeri {1}'.format(y[Pear_Ind[0, i]],1-Pear_Deger[0,i],i+1))

    
```
